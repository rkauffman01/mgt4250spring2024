```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("student-mat.csv")

```

```python
X = df[['age','Medu','Fedu',
        'traveltime','studytime','failures','famrel','freetime','goout','Dalc','Walc','health','absences']]
y = df['G3']

X2 = df[['age','Medu','Fedu',
        'traveltime','studytime','failures','famrel','freetime','goout','health','absences']]
y2 = df['Walc']


X3 = df[['age','Medu','Fedu',
        'traveltime','studytime','failures','famrel','freetime','goout','health','absences']]
y3 = df['Dalc']
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=13)


from sklearn import tree

model = tree.DecisionTreeClassifier(max_depth=2)
model.fit(Xtrain, ytrain)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
class_names = [str(cls) for cls in ytrain.unique()]
tree.plot_tree(model, feature_names=X.columns,  
               class_names=class_names,
               rounded=True, filled=True, ax=ax)


```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X2, y2, test_size=0.25, random_state=13)

model = tree.DecisionTreeClassifier(max_depth=2)
model.fit(Xtrain, ytrain)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
class_names = [str(cls) for cls in ytrain.unique()]
tree.plot_tree(model, feature_names=X2.columns,  
               class_names=class_names,
               rounded=True, filled=True, ax=ax)
```
